Це завдання - на продуктове мислення. Якщо не знаєте, які гіпотези сформулювати - напишіть ті, які здаються вам зовсім примітивними. Важливо навчитись висловлювати свої гіпотези щодо даних, і з часом та досвідом вони будуть ставати щоразу глибше і точніше. Весь аналіз даних будується на тому, що ми перевіряємо певні гіпотези та робимо висновки і рекоемндації на основі своїх досліджень.

```python
yangon_mean = df[df.City =='Yangon']['gross income'].mean()
mandalay_mean = df[df.City =='Mandalay']['gross income'].mean()

print ('Yangon', '=', yangon_mean)
print ('Mandalay', '=', mandalay_mean)
```

У місті Mandalay середнє значення gross income більше. Це може бути спричинене наступними факторами:
1. У  Mandalay вищий середній чек, що може бути, наприклад, за рахунок більшої кількості проданих товарів або за рахунок того, що покупці купують дорожчі товари тощо. Тут ми можемо додатково порівняти середню кількість проданих товарів в інвойсі по кожному місту, середню ціну одиниці товару. Для більш детального аналізу можна також проаналізувати купівельну спроможність населення, середній дохід домогосподарств, демографічну ситуацію в цих містах, якщо є такі дані.
2. Також на gross income може впливати різниця у типі клієнтів (Member vs Normal) та їх поведінка. Тому потрібно дослідити, яка частка Member клієнтів у кожному місті, і який вони дають дохід, якщо у Mandalay частка Member клієнтів більша, ніж у Yangon, то слід зосередитися на її підвищенні у цьому місті за допомогою різних методів залучення (масштабувати програму лояльності в Yangon, впровадити персоналізовані пропозиції для Members, можливо запровадити якісь додаткові бонуси на покупку від якоїсь певної суми тощо).
